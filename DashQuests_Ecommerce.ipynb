```python
import pandas as pd
file_path = '/content/drive/MyDrive/DashQuest_Guvi/Amazon Sale Report - Amazon Sale Report.csv.csv'  # or .csv
df = pd.read_csv(file_path)
```

```python
# ✅ 5. Check for nulls and handle them
print(df.isnull().sum())
```

```python
# Fill or drop missing data as needed
df['Courier Status'] = df['Courier Status'].fillna('Unknown')
df.dropna(subset=['Amount', 'Date'], inplace=True)
```

```python
df.drop(columns=['promotion-ids', 'fulfilled-by', 'Unnamed: 22'], inplace=True)
```

```python
df.drop(columns=['ship-city', 'ship-state', 'ship-postal-code', 'ship-country'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# ✅ Convert data types (if not done yet)
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce')
df['B2B'] = df['B2B'].astype(bool)
```

```python
print(df.isnull().sum())
```

```python
# ✅ Drop any rows with missing critical sales values
df.dropna(subset=['Amount', 'Qty', 'Date', 'currency'], inplace=True)

# ✅ OPTIONAL: Fill Qty if you do not want to drop rows
df['Qty'] = df['Qty'].fillna(df['Qty'].median())

# ✅ Feature Engineering
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.strftime('%b')
df['Total_Sale'] = df['Qty'] * df['Amount']
```

```python
df.head(2)
```

```python
# ✅ Export cleaned data to CSV (for Power BI)
output_path = '/content/drive/MyDrive/DashQuest_Guvi/cleaned_amazon_sales.csv'
df.to_csv(output_path, index=False)
print(f"✅ Cleaned CSV saved: {output_path}")
```
